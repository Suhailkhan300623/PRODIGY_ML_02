```python
from sklearn.cluster import KMeans
```

```python
x=data.loc[:,['Age','Spending Score (1-100)']].values
wcss=[]

for i in range(1,11):
    k_means=KMeans(n_clusters=i)
    k_means.fit(x)
    wcss.append(k_means.inertia_)
    
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color='red', marker='*')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4)
label=kmeans.fit_predict(x)
label
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1], c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.title('Customers')
plt.xlabel('Age')
plt.ylabel('Spending')
plt.show()
```

```python
X=data.iloc[:,1:].values
```

```python
WCSS=[]

for i in range(1,11):
    kMeans=KMeans(n_clusters=i)
    kMeans.fit(X)
    WCSS.append(kMeans.inertia_)
```

```python
plt.plot(range(1,11), WCSS, linewidth=2, marker='o', color='red')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow')
plt.show()
```

```python
from sklearn.cluster import KMeans
KMeans=KMeans(n_clusters=6)
Labels=KMeans.fit_predict(X)
```

```python
data['Label']=Labels
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=Labels, cmap='rainbow')
centers = KMeans.cluster_centers_
ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], marker='x', c='black', s=100)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending')

plt.show()
```
